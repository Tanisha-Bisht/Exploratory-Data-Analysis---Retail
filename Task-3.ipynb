## <b>Bivariate analysis of profit column with the rest of the columns</b>

```python
# Comparision of profit regionwise

plt.figure(figsize=(10,7))
data_group=data.groupby(["Region"])["Profit"].sum()
data_group.sort_values(ascending=False)
data_group.plot.bar()
plt.xticks(rotation=360)
plt.show()
```

```python
# Comparision of profit citywise

plt.figure(figsize=(10,10))
city_profit=data.groupby(["City"])["Profit"].sum()
a=city_profit[city_profit>3000]
a.sort_values(ascending=False)
a.plot.bar()
plt.xticks(rotation=70)
plt.show()
```

```python
#Comparision of profit statewise

plt.figure(figsize=(20,10))
data_group=data.groupby(["State"])["Profit"].sum()
data_group.sort_values(ascending=False)
data_group.plot.bar()
plt.xticks(rotation=90)
plt.show()
```

#### <u><b>Note: States like California and New York are giving us more profit and states like Texas and Ohio are giving us less profit</b></u>

```python
#Comparision of profit respective to sales

plt.figure(figsize=(10,5))
sb.scatterplot(data=data, x="Sales", y="Profit")
plt.show()
```

#### <u><b>Here we can observe that sale and profit are not depending factor to each other , thus sale is not playing a vital role in increasing the profit </u></b>

```python
#pairplot
sb.pairplot(data)
plt.show()
```
